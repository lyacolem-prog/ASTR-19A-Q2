#### Imports Modules Required

```python
from tabulate import tabulate 
import numpy as np
```

#### Defines sin_x function

```python
def sin_x(x):
    return np.sin(x)
```

#### Defines cos_x function

```python
def cos_x(x):
    return np.cos(x)
```

#### Generates List using Numpy, creates lists for each axis and appends 10 values to each, then uses tabulate to create a table for all of them, then prints that table. 

```python
x = np.linspace(0, (np.pi*2), 1000) # Generates List data using Numpy


table_data = zip(x, sin_x(x), cos_x(x)) # Creates Table_data directly from Numpy List so there are 1000 Entries
table_headers = ["x", "sin(x)", "cos(x)"] # Creates Table Headers
table = tabulate(table_data, tablefmt="grid", headers=table_headers, floatfmt=".3f") # creates table from table_data



# Creates sin_axis list and appends 10 values to it
sin_axis = []
for i in range(10):
    sin_axis.append(sin_x(x)[i])

# Creates cos_axis list and appends 10 values to it
cos_axis = []
for i in range(10):
    cos_axis.append(cos_x(x)[i])


table_data_10 = zip(x, sin_axis, cos_axis) # Creates Table_data from axis lists so there is only 10 entries
table_10 = tabulate(table_data_10, tablefmt="grid", headers=table_headers, floatfmt=".3f") # creates table from table_data

# prints out the table
print(table_10)


```

```python
x = np.linspace(0, (np.pi*2), 1000) # Generates List data using Numpy

# Creates sin_axis list and appends 10 values to it
sin_axis = []
for i in range(10):
    sin_axis.append(sin_x(x)[i])

# Creates cos_axis list and appends 10 values to it
cos_axis = []
for i in range(10):
    cos_axis.append(cos_x(x)[i])


table_data = zip(x, sin_axis, cos_axis) # Creates Table_data from axis lists
table_headers = ["x", "sin(x)", "cos(x)"] #Creates Table Headers
table = tabulate(table_data, tablefmt="grid", headers=table_headers, floatfmt=".3f") # creates table from table_data

# prints out the table
print(table)

```
