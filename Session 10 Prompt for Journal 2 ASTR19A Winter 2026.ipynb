```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Generates fake data in a triangular fashion
x = np.random.triangular(0,1,2,1000)

# Creates the Histrogram 
plt.hist(x,bins=100,alpha=0.9,edgecolor='red')
plt.xlabel('x')
plt.ylabel('number per bin')

plt.show() # I added the plt.show without thinking... im not sure if im supposed to include it!
```
