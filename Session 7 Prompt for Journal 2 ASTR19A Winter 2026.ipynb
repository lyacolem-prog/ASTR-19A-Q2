```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generates Fake data
x = np.linspace(0,1,100)
```

```python
# Creates Exponent Function
def exp(x):
    y = x**x
    return y

# Defins the plot under the name "figure"
figure = plt.plot(x,exp(x))

# Sets Labels for the plot
plt.xlabel('Time [miliseconds]')
plt.ylabel('Awesomeness')
```

```python
plt.savefig('plot.pdf', format ="pdf", bbox_inches='tight',dpi=600)
```
