```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generates fake data
x = np.linspace(0,1,100)

# Defines Functions sin(x) and cos(x)
def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)
```

```python
sinx = np.sin(x)
cosx = np.cos(x)

figure, axes = plt.subplots(1,2)

# first plot
axes[0].plot(x,sinx)
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
axes[0].set_title('sin(x) from 0 to 1')

# second plot
axes[1].plot(x,cosx)
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')
axes[1].set_title('cos(x) from 0 to 1')
```

```python
plt.savefig('Sinx_vs_cosx.pdf', format ="pdf", bbox_inches='tight',dpi=600)
```
