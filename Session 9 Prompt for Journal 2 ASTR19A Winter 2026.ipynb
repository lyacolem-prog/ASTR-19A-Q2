```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Generates fake data
x = np.random.uniform(0,1,1000)

# creates constraints for the bins 
width = 0.04 # width of bins
histmin = np.floor(min(x))
histmax = np.ceil(max(x))

# Creates the bins
bins = np.arange(histmin,histmax,width)

# Creates the Histrogram 
plt.hist(x,bins=bins,alpha=0.9,edgecolor='yellow')
plt.xlabel('x')
plt.ylabel('number per bin')

plt.show() # I added the plt.show without thinking... im not sure if im supposed to include it!
```
